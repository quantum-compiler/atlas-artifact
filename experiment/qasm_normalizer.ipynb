```python
# normalize a circuit from QpenQASM 3.0 to Quartz readable qasm
# Gate set: rigetti

import os
import re
from os.path import join, isfile
```

```python
origin_dir = os.getcwd() + '/circs/ibm_circs'
output_dir = os.getcwd() + '/circs/ibm_special_angle_circs'
fns = os.listdir(origin_dir)
for fn in fns:
    with open(origin_dir + '/' + fn, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    with open(output_dir + '/' + fn, 'w') as of:
        for line in lines:
            if line[:2] != 'rz':
                new_lines.append(line)
            elif line.find("rz(-pi/4)") != -1:
                new_lines.append(line.replace("rz(-pi/4)", "tdg"))
            elif line.find("rz(pi/4)") != -1:
                new_lines.append(line.replace("rz(pi/4)", "t"))
            elif line.find("rz(-pi/2)") != -1:
                new_lines.append(line.replace("rz(-pi/2)", "tdg"))
                new_lines.append(line.replace("rz(-pi/2)", "tdg"))
            elif line.find("rz(pi/2)") != -1:
                new_lines.append(line.replace("rz(pi/2)", "t"))
                new_lines.append(line.replace("rz(pi/2)", "t"))
            elif line.find("rz(-3*pi/4)") != -1:
                new_lines.append(line.replace("rz(-3*pi/4)", "tdg"))
                new_lines.append(line.replace("rz(-3*pi/4)", "tdg"))
                new_lines.append(line.replace("rz(-3*pi/4)", "tdg"))
            elif line.find("rz(3*pi/4)") != -1:
                new_lines.append(line.replace("rz(3*pi/4)", "t"))
                new_lines.append(line.replace("rz(3*pi/4)", "t"))
                new_lines.append(line.replace("rz(3*pi/4)", "t"))
            elif line.find("rz(-pi)") != -1:
                new_lines.append(line.replace("rz(-pi)", "z"))
            elif line.find("rz(pi)") != -1:
                new_lines.append(line.replace("rz(pi)", "z"))
            else:
                print(line)

        of.writelines(new_lines)
    
```

```python
origin_dir = os.getcwd() + '/circs/ibm_extra_circs'
output_dir = os.getcwd() + '/circs/ibm_special_angle_extra_circs'
fns = os.listdir(origin_dir)
for fn in fns:
    with open(origin_dir + '/' + fn, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    with open(output_dir + '/' + fn, 'w') as of:
        for line in lines:
            if line[:2] != 'rz':
                new_lines.append(line)
            elif re.findall(r"[0-9]*\.5\*pi", line) != []:
                f = re.findall(r"[0-9]*\.5", line)[0]
                new_f = float(f)
                while new_f > 2:
                    new_f -= 2
                while new_f < 0:
                    new_f += 2
                if str(new_f) == "0.5":
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
                elif str(new_f) == "1.5":
                    new_lines.append(line.replace(f"rz({f}*pi)", "z"))
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
            elif re.findall(r"[0-9]*\.0\*pi", line) != []:
                f = re.findall(r"[0-9]*\.0", line)[0]
                new_f = float(f)
                while new_f > 2:
                    new_f -= 2
                while new_f < 0:
                    new_f += 2
                if str(new_f) == "1.0":
                    new_lines.append(line.replace(f"rz({f}*pi)", "z"))
                else:
                    continue
            elif re.findall(r"[0-9]*\.25\*pi", line) != []:
                f = re.findall(r"[0-9]*\.25", line)[0]
                new_f = float(f)
                while new_f > 2:
                    new_f -= 2
                while new_f < 0:
                    new_f += 2
                if str(new_f) == "0.25":
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
                elif str(new_f) == "1.25":
                    new_lines.append(line.replace(f"rz({f}*pi)", "z"))
                    new_lines.append(line.replace(f"rz({f}*pi)", "t"))
            elif re.findall(r"[0-9]*\.75\*pi", line) != []:
                f = re.findall(r"[0-9]*\.75", line)[0]
                new_f = float(f)
                while new_f > 2:
                    new_f -= 2
                while new_f < 0:
                    new_f += 2
                if str(new_f) == "0.75":
                    new_lines.append(line.replace(f"rz({f}*pi)", "tdg"))
                elif str(new_f) == "1.75":
                    new_lines.append(line.replace(f"rz({f}*pi)", "z"))
                    new_lines.append(line.replace(f"rz({f}*pi)", "tdg"))
            else:
                print(line)
                pass


        of.writelines(new_lines)
    
```

```python
qasm_path = os.getcwd() + '/rigetti_circs/'
qasm_fns = [fn for fn in os.listdir(qasm_path) if isfile(join(qasm_path, fn)) and fn[-4:] == 'qasm' and fn[0] != '_']
qasm_full_paths = [qasm_path + fn for fn in qasm_fns]
```

```python
for qasm_fn in qasm_full_paths:
    with open(qasm_fn, 'r') as f:
        text: str = f.read()
        lines: list[str] = text.split('\n')
        lines = [line for line in lines if line != '' and line[:2] != '//']
        lines = [line for line in lines if line[:7] != 'barrier' and line[:7] != 'measure']
        lines = [line for line in lines if line[:7] != 'rz(0.0)']

        new_lines: list[str] = []
        for line in lines:
            line += '\n'
            if line[:2] == "rx":
                if line[:8] == "rx(pi/2)":
                    new_lines.append(line.replace("rx(pi/2)", "rx1"))
                elif line[:6] == "rx(pi)":
                    new_lines.append(line.replace("rx(pi)", "x"))
            else:
                new_lines.append(line)

    with open(f"{qasm_fn[:-5]}_norm.qasm", 'w') as f:
        f.writelines(new_lines)
```

```python
import quartz

quartz_context = quartz.QuartzContext(gate_set=['cz', 'x', 'rz', 'add', 'rx1', 'rx3'], filename='ecc_set/rigetti_4_ecc.json')
for qasm_fn in qasm_full_paths:
    print(qasm_fn)
    if qasm_fn.find("norm") == -1:
        continue
    circ = quartz.PyGraph.from_qasm(context=quartz_context, filename=qasm_fn)
    print(circ.to_qasm_str())
```
